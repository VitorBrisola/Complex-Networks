```python
# reading txt
G_lm = nx.read_edgelist('data/lesmis.txt', nodetype=int, data=(('weight',float),),create_using= nx.DiGraph())
```

```python
# reading gml
#G_lm = nx.read_gml('data/lesmis.gml', weight='value')
```

__Drawing__ the _Les_ _Miserables_ __directed__ and __weighted__ network version.
> The edges __weights__ are represented by its widths

```python
# weights setting 
edges = G_lm.edges()
weights = [G_lm[u][v]['weight'] for u,v in edges]
```

```python
# printing
pos=nx.spring_layout(G_lm)
nx.draw_networkx(G_lm, with_labels = False, node_color='red', edge_color='black', 
        node_size=50, font_size=16, pos=pos, width=weights)
plt.show(True)
```

__Drawing__ the _Les_ _Miserables_ __undirected__ and __unweighted__ network version.

```python
# transforming network to undirected
G_lm = G_lm.to_undirected()
```
