```python
# printing network
def print_net(N,with_labels=False,node_size=50,font_size= 16,label=None): 
    if with_labels:
        labels = N.nodes()
    pos=nx.spring_layout(N)
    nx.draw(N, with_labels =  with_labels, node_size=node_size, font_size=font_size,pos=pos,label=label)
    plt.show(True)
```

### Degree

```python
def get_degrees(G):
    vk = dict(G.degree())
    vk = list(vk.values())
    degree = np.array(vk)
    return degree
```

#### __Degree Distribution__

```python
# calc the degree distribution
def get_degree_distribution(G):
    vk = get_degrees(G)
    maxk = np.max(vk)
    mink = np.min(vk)
    kvalues= arange(0,maxk+1) # possible values of k
    Pk = np.zeros(maxk+1) # P(k)
    for k in vk:
        Pk[k] = Pk[k] + 1
    Pk = Pk/sum(Pk) # the sum of the elements of P(k) must to be equal to one
    return kvalues,Pk
```

```python
# plot degree distribution
def plot_degree_distribution(Ks, Pk,plot_type=None):    
    fig = plt.subplot(1,1,1)
    #fig.set_xscale('log')
    #fig.set_yscale('log')
    
    # selecionando o tipo de plot('histogram','scatterplot',..)
    if plot_type == 'bo':
        plt.plot(Ks,Pk,'bo')
    #elif plot_type == 'hist':
    #    plt.hist(Ks,Pk, bins  = len(ks))
    else:
        plt.plot(Ks,Pk)
        
    plt.xlabel("k", fontsize=20)
    plt.ylabel("P(k)", fontsize=20)
    plt.title("Degree distribution", fontsize=20)
    #plt.grid(True)
    #plt.savefig('degree_dist.eps') #save the figure into a file
    plt.show(True)
```

```python
def momment_of_degree_distribution(G,m):
    k,Pk = get_degree_distribution(G)
    M = sum((k**m)*Pk)
    return M
```

```python
def momment_of_degree_distribution2(G,m):
    M = 0
    N = len(G)
    for i in G.nodes:
        M = M + G.degree(i)**m
    M = M/N
    return M
```

```python
def shannon_entropy(G):
    k,Pk = get_degree_distribution(G)
    H = 0
    for p in Pk:
        if(p > 0):
            H = H - p*math.log(p, 2)
    return H
```

```python
def normalized_shannon_entropy(G):
    k,Pk = get_degree_distribution(G)
    H = 0
    for p in Pk:
        if(p > 0):
            H = H - p*math.log(p, 2)
    return H/math.log(len(G),2)
```

### Probability Matrix

_Rotina para calcular a matrix de probabilidade de uma data rede._

```python
def probability_matrix(G):
    adj_M = nx.to_numpy_matrix(G) # adjacency matrix
    degrees = get_degrees(G)
    
    for node_idx in range(len(degrees)):
        node = adj_M[node_idx]
        new_node = np.divide(node,degrees[node_idx])
        adj_M[node_idx] = new_node
    
    prob_M = adj_M
    return prob_M
```

### Distance 

```python
def distance_matrix(G):
    N = len(G.nodes)
    D = zeros(shape=(N,N)) # D is the matrix of distances
    vl = []
    for i in arange(0,N):
        for j in arange(i+1, N):
            if(i != j):
                aux = nx.shortest_path(G,str(i+1),str(j+1))
                dij = len(aux)-1
                D[i][j] = dij
                D[j][i] = dij
                vl.append(dij)
    #x = range(0,d+1)
    
    return D
```

#### __Distribution of the shortest paths__ 

```python
def shortest_paths_distribution(G):
    if nx.is_connected(G) == True:
        N = len(G.nodes)
        D = zeros(shape=(N,N)) # D is the matrix of distances
        vl = []
        # getting distance
        for i in arange(0,N):
            for j in arange(i+1, N):
                if(i != j):
                    aux = nx.shortest_path(G,i,j)
                    dij = len(aux)-1
                    D[i][j] = dij
                    D[j][i] = dij
                    vl.append(dij)
        d = nx.diameter(G)
        x = range(0,d+1)
        plt.hist(vl, bins = x)
        plt.title("Distribution of the Shortest Paths", fontsize=20)
        plt.ylabel("P(l)", fontsize=15)
        plt.xlabel("Shortest path length (l)", fontsize=15)
        #plt.grid(True)
        plt.savefig('av_short_path.eps')
        plt.show(True)
    else:
        print("The graph has more than one connected component")
```

### Random Walk

_A rotina abaixo simula uma caminhada aleatória pela rede e retorna uma lista com o número de vizitas recebidos por cada nó._

```python
def random_walk(G,l):
    nv = np.zeros(len(G.nodes)+1) # visits counter for each node
    for node in G.nodes:
        exit = False
        num_steps = l
        pos = node
        while(not exit and num_steps > 0):
            neighbors = list(G.neighbors(pos))
            if(len(neighbors) > 0):
                next_node = neighbors[rand.randint(0,len(neighbors)-1)]
                nv[next_node] += 1 # visit counting
                pos = next_node
                num_steps -= 1 
            else:
                exit = True
                
    return nv
```

### Components

__Largest Component__ extracting

```python
def get_largest_component(G,show = False):
    Gcc=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)
    G=Gcc[0]
    # drawing largest component
    if show:
        print_net(G)
    return G
```

```python
# carregando matrizes de probabilidade
powers = [10,100,1000]
pm = dict()
pm['1'] = probability_matrix(G_ex)
for power in powers:
    pm[power] = matrix_power(pm['1'],power)
    # vizualizando matrizes resultantes
    #print("=== Probability Matrix Powered by "+str(power)+" ===")
    #print(pm[power])
    #print('\n')
```

```python
# ploting transpose matrix
fig, axs = plt.subplots(2, 2,figsize=(15,8))

for ax,key in zip(axs.flat,pm):
    ax.plot(pm[key].transpose())
    ax.set_title('Probability Matrix Powered by '+str(key))
    ax.set(xlabel='node', ylabel='(node)')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

>__Transpose Matrix Ploting:__ A matriz transposta da matriz de probabilidade coloca a probabilidade deum nó ser percorrido iniciado-se em um determinado nó, dado agora pela coluna. Logo a figura acima mostra como a probabilidade de um nó ser percorrido torna-se independente do nó inicial .

### __Table__ construction

```python
def get_table(G):
    table = [['Number of nodes',len(G)],
            ['Average degree',np.mean(get_degrees(G))],
            ['second moment of degree distribution',momment_of_degree_distribution(G,2)],
            ['average clustering coefficient',nx.average_clustering(G)],
            ['transitivity',nx.transitivity(G)],
            [' average shortest path length',nx.average_shortest_path_length(G)],
            [' diameter',nx.diameter(G)]] 
    return np.array(table)
```

```python
def table_plot(table_data,labels,title=None):
    plt.figure(figsize=(7,3))
    if title: plt.title(title, fontsize=20)
    plt.axis('tight')
    plt.axis('off')
    ytable = plt.table(cellText=table_data,colLabels=labels,cellLoc='right',loc='best')
    ytable.set_fontsize(34)
    ytable.scale(1, 2)
    plt.show()
```
