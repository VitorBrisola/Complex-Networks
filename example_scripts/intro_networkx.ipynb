```python
# importing networkx library
import networkx as nx
```

```python
# Creating a new graph to use as a network
G = nx.Graph()
```

```python
# adding graph edges
G.add_edge('a','b',weight=0.1)
G.add_edge('a','c',weight=0.5)
G.add_edge('b','c',weight=0.3)
G.add_edge('a','d',weight=0.9)
```

```python
# Using matplot lib to visualize the graph
import matplotlib.pyplot as plt

labels = G.nodes()
pos=nx.spring_layout(G)
nx.draw_networkx(G, with_labels = True, nodecolor='r', edge_color='b', 
        node_size=500, font_size=16, pos=pos, width=6)   # default spring_layout
# use default edge labels
nx.draw_networkx_edge_labels(G,pos)
plt.savefig('graph.pdf')
plt.show(True)
```

```python
# Joining two graphs
## graph 0
G0 = nx.Graph([(0,1),(1,2),(2,0)])
nx.draw(G0, with_labels = True, node_size=500, font_size=16)
plt.show(True)
```

```python
## Graph 1
G1 = nx.Graph([(0,1),(1,2),(2,3), (3,4), (2,4)])
nx.draw(G1, with_labels = True, node_size=500, font_size=16)
plt.show(True)
```

```python
# joining graphs using disjoint_union that assumes all nodes are different
# graph 2 = g0 union g1
G2 = nx.disjoint_union(G0,G1)
nx.draw(G2, with_labels = True, node_size=500, font_size=16)
plt.show(True)
```

```python
# graph now has 7 nodes
print('Nodes:', G2.nodes())
print('Edges:', G2.edges())
```

```python
# joining graphs by composing where nodes with the same labels are seen as the same
G3 = nx.compose(G0,G1)
nx.draw(G3, with_labels = True, node_size=500, font_size=16)
plt.show(True)
```

```python
# Graph Conversion
# Converting directed graph to an undirected one
## Directed graph
G.clear()
edgelist=[(0,1),(1,2),(2,3), (3,1), (4,1)]
G = nx.DiGraph(edgelist)
pos=nx.spring_layout(G)
nx.draw(G, with_labels = True, node_size=500, font_size=16, pos = pos)
plt.show(True)
```

```python
## Undirected graph conversion
G = G.to_undirected()
nx.draw(G, with_labels = True, node_size=500, font_size=16, pos = pos)
plt.show(True)
```

```python
# converting an undirected to direct
# the new graph will have both aways direction
G = G.to_directed()
nx.draw(G, with_labels = True, node_size=500, font_size=16, pos = pos)
plt.show(True)
```

```python
# Converting weightened network to an unweighted

## Weightened graph
G1.clear()
G1 = nx.Graph()
G1.add_edge(1, 2, weight=4.0 )
G1.add_edge(1, 3, weight=2.0 )
G1.add_edge(2, 3, weight=1.0 )
labels = G1.nodes()
pos=nx.circular_layout(G1)
nx.draw(G1, with_labels = True, nodecolor='r', edge_color='b', 
        node_size=500, font_size=16, pos=pos, width=6)   # default spring_layout
# use default edge labels
nx.draw_networkx_edge_labels(G1,pos)
plt.show(True)
```

```python
# Removing edges that have lower weight than a threshold
# Storing removed edges in a new unweightened graph
min_weight = 1.0
Gnew = nx.Graph()
for (u,v,w) in G1.edges(data=True):
    if w['weight'] > min_weight :
        Gnew.add_edge(u, v, weight=1)
        
```

```python
labels = Gnew.nodes()
nx.draw(Gnew, with_labels = True, nodecolor='r', edge_color='b', 
        node_size=500, font_size=16, pos=pos, width=6)   # default spring_layout
# use default edge labels
nx.draw_networkx_edge_labels(Gnew,pos)
plt.show(True)
```

```python
# Components
# the following network present 3 components
G.clear()
G = nx.Graph([(0,1),(1,2),(2,3), (3,4), (2,4), (5,6), (6,7), (8,9)])
pos = nx.fruchterman_reingold_layout(G);
nx.draw(G, with_labels = True, node_size=500, font_size=16, pos = pos)
plt.show(True)
```

```python
# getting components
Gcc=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)
print(Gcc[0].nodes())
print(Gcc[1].nodes())
print(Gcc[2].nodes())
```

```python
# Getting largest component
Gcc=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)
G=Gcc[0]
print(G.nodes())
pos = nx.fruchterman_reingold_layout(G);
nx.draw(G, pos = pos, with_labels = True)
plt.show()
```

```python
# Removing edges while keeping nodes
#Edge removal
import numpy as np
import random as random
def remove_node_edges(G, i):
    ng = G.neighbors(i)
    l = []
    for j in ng:
        e = [i,j]
        l.append(e)
    G.remove_edges_from(l)
    
N = 10
p = 0.2
G = nx.gnp_random_graph(N, p, seed=None, directed=False)

Gcc=sorted(nx.connected_component_subgraphs(G), key = len, reverse=True)
G=Gcc[0]
print(G.nodes())
pos = nx.fruchterman_reingold_layout(G);
nx.draw(G, pos = pos, with_labels = True)
plt.show()
```

```python
# removing nodes

node = random.choice(np.array(G.nodes()))
print('Node removed:', node)
remove_node_edges(G, node)
nx.draw(G, pos = pos, with_labels = True)
plt.show()
print(G.nodes())
```
